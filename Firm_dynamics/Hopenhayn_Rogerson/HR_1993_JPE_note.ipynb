## Abstract

A lot of empirical works suggest the number of job creation and destruction **at the level of indivisual firm** is very large.
However, the aggregate variables does not reflect them since they are cancelled out.

This paper builds general equilibrium model with heterogenous firms and consider the aggregate effects of policy that interfere
 the reallocation at the level of indivisual firm. In particular, they pay attension to dismissal cost.
 
 

## Model

**Environment**

Markets are complete and perfectly competitive. No saving i.e. no capital.

There are representative household and heterogeneous firms.

We consider stationary equilibrium where aggregate variables and prices are constant.

**Firm**

Each firm recieves profit at time t given by
$$p_t f(n_t,s_t) -n_t -p_t c_f -g(n_t, n_{t-1})$$

- $p_t$:Output price

- wage: Normalized to 1. 

- $c_f$  Fixed operating cost. Denominated in units of output

- $f(n_t,s_t)$:Stochastic production function. 

- $n_t$: Employment. 

- $s_t$:Stochastic productivity which follows a first order Markov process. 

- $g(n_t, n_{t-1})$: Employment adjustment cost

**Timing of decision - Incumbents**


Incumbents that works in time $t-1$ has an information about $s_{t-1}, n_{t-1}$. 

At the beginning of time $t$,
 they decide to exit or keep working in this period. 
 
 If they exist,they have to pay $g(0,n_{t-1})$, since they do not employ anyone in this period.
 
 If they stay, they recieve $s_t$ and decide how many people they hire.
 
 Then incumbents recieve profit and repeat this procedure in the next period.


**Timing of decidion - Entrants**

Assume there exist a lot of potential entrants.

If they enter, pay entry cost  $c_e$ (Denominated in units of output), then recieve stochastic productivity as a draw from the distribution $\nu$.

After the entrance, their situation is same as incumbents.




**Bellman Equation **

Bellman equation of the firm can be written as follows. ($n'$ is today's employment.)

$$W(s,n;p) = \max_{n'\ge 0}\left\{pf(n',s) - n' - pc_f - g(n',n) + \beta \max\left\{ E_s\left[W(s',n';p), -g(0,n')\right]\right\}\right\}$$

- $\beta$: discount factor. same as $\frac{1}{1+r}$


If $W$ is known, the value of the entrant before sutracting fixed entry cost is given by

$$W^e(p) = \int W(s,0;p)d\nu(s)$$

Let policy functions be 
$$
N(s,n,p)
$$

$$
X(s,n,p) 
$$

The former is a policy function for employment and the latter is a policy function for entry-exit choice.
$X(s,n,p) =1$ corresponds to exit.


**Measure of firms **

Let a measure over pairs (s,n) be $\mu(s,n)$ and let the mass of entrants be $M$.

For each period, measure changes from $\mu$ to $\mu'$ after the decision whether exit or not and recieving new producivity information.

Define $T(\mu, M;p)$ that captures this transition.

**Aggregation**

Using previous settings, we can compute output, labor demand and profit.

$$Y(\mu,M;p) = \int\left[f(N(s,n;p),s) - c_f\right]d\mu(s,n) + M\int\left[f(N(s,0;p),s)\right]d\nu(s)$$

*Note *

Assume new entrants bears fixed entry cost but not operation cost. After the entry period, they become incumbents and must pay 
operation cost.

$$L^d(\mu,M,p) = \int N(s,n;p)d\mu(s,n) + M\int N(s,0;p)d\nu(s)$$

To compute aggregate profit, we need to expected adjustment costs for the next period at each firm. It is given by

$$r(s,n;p) = [1-X(s,n;p)]\int g(N(s',n';p),n')dF(s,s') + X(s,n;p)g(0,n')$$

Let aggregate adjustment cost be $R(\mu,M,p)$ which is defined by integration.

$$\Pi(\mu,M,p) = pY(\mu,M,p) -L^d(\mu,M,p) -R(\mu,M,p) - Mpc_e$$

**Household**

We consider Hansen(1985) style household. (If you don't know it, see Hansen's paper.)

Representative household's discount expected life time utility is given by

$$\sum^{\infty}_{t=1} \beta^t\left\{ u(c_t) - aN_t\right\}$$

$N_t$ is the fraction of indivisuals who are employed at time t. a is some constant. Assume $c_t $ is nonnegative.

Since the economy does not have asset, we can redure their maximization problem to a static optimization problem like

$$\max u(c) - aN$$
subject to 
$$pc \le N + \Pi +R$$

where $\Pi$ is profit and $R$ is tax receipt.

Solving this problem, we obtain labor supply function $L^s(p, \Pi+R)$. 

## Definition of Equilibrium

\begin{definition} Stationary equilibrium

A stationary equilibrium  consists of an output price $p^\ast \ge 0$, a mass of entrants $M^\ast \ge 0$, and 
a measure of incumbents $\mu^\ast$, 

such that

i) Labor market clears

$L^d(\mu^\ast, M^\ast, p^\ast) = L^s(p^\ast, \Pi(\mu^\ast,M^\ast,p^\ast)+R(\mu^\ast,M^\ast,p^\ast))$

ii) There exists a invariant distribution over firms

$T(\mu^\ast,M^\ast;p^\ast) = \mu^\ast$

iii) Free entry condition

$W^e(p^\ast) \le p^\ast c_e$ with equality if $M^\ast >0$
\end{definition}

## Algorithm

The way to find stationary equilibrium consists of two steps.

** Step1. Iterate over $p$ until the entry condition is satisfied at $p^\ast$ **

- Guess output price, $p_i$

- Let $EC(p_i) = \frac{\int W^e(s;p_i)}{p_i} -c^e $
 
- If $EC(p_i)>0$, let $p_{i+1} < p_{i}$. Otherwise let $p_{i+1} > p_{i}$. go back to computeing $EC(p_i)$.
 
- If $EC(p_i)$ is sufficiently close to 0, stop iteration.
 
**Step2. Iterate over$(\mu_i, M_i)$ until labor market clears at $(\mu^\ast, M^\ast)$**

- Let $M_0=1$ Find a fixed point $\mu^{ss}_0(M_0 = 1)$ using transition function of distribution.

- Let $LMC(\mu_i,M_i) = L^d(\mu^{ss}_i, M_i) - N^s(p^\ast,\Pi(\mu^{ss}_i(M_i) ),M_i; p^\ast)$.

- If $LMC(\mu_i,M_i) > 0$, then set $ M_{i+1} < M_i$. Otherwise set $M_{i+1} > M_i$. Go back to find a fixed point.

- If $LMC(\mu_i,M_i)$ is sufficiently close to 0, stop iteration.
